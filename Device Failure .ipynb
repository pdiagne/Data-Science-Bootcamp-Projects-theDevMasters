```python
#Prepare data for number of days until failure calc

#Reload data
import pandas as pd
DatA = pd.read_csv('data.csv')

#Convert date
Dates = pd.DataFrame()
Dates = pd.to_datetime(DatA.date)
Date_conv = pd.DataFrame({'year':Dates.dt.year,'month':Dates.dt.month,'day':Dates.dt.day})
DatA = DatA.join(Date_conv)

#Last entry
DatA_L = DatA.sort_values(by=['device','date'],ascending = False)
DatA_L = DatA_L.drop_duplicates(subset=['device'],keep = 'first')
DatA_L = DatA_L.reset_index(drop=True)

#First entry
DatA_F = DatA.sort_values(by=['device','date'],ascending = True)
DatA_F = DatA_F.drop_duplicates(subset=['device'],keep = 'first')
DatA_F = DatA_F.sort_values(by=['device'],ascending = False)
DatA_F = DatA_F.reset_index(drop=True)

#print(len(DatA_L)) == print(len(DatA_F)) #True (1168)
```

```python
print(len(DatA_F))
DatA_F.head()
```

```python
print(len(DatA_L))
DatA_L.head()
```

```python
#Caluclate average days in service/days until failure
time = pd.DataFrame()
DatA_L.failure = DatA_L.failure.astype(str)

for i in range(len(DatA_L)):
    if DatA_L.get_value(i,col='failure') == '1':
        dFday = abs(DatA_F.get_value(i,col='day') - DatA_L.get_value(i,col='day'))
        dFmonth = 30*abs(DatA_F.get_value(i,col='month') - DatA_L.get_value(i,col='month'))
        time.set_value(i,'days',dFday+dFmonth) 
time_avg = time.mean() #104 average days until failure
time_avg
```

# Baseline Model

```python
#Convert categorical int to str 
data.attribute3 = data.attribute3.astype(str)
data.attribute5 = data.attribute5.astype(str)
data.attribute7 = data.attribute7.astype(str)
data.attribute9 = data.attribute9.astype(str)

#Convert failure int to str
data.failure = data.failure.astype(str)
data.info()
```

```python
#Replace top end outliers with median
x_var_list = ['attribute1','attribute2','attribute4','attribute6']
for g in range(0,len(x_var_list)):
    A = data[x_var_list[g]].std()
    C = data[x_var_list[g]].mean()
    D = data[x_var_list[g]].median()
    for r in range(0,len(data[x_var_list[g]])):
        if data[x_var_list[g]].get_value(r)-C >= 3*A:
            data.set_value(r,[x_var_list[g]],D)
```

```python
#Dummy Categoricals
dum_list = ['attribute3','attribute5','attribute7','attribute9']
merged_data_dums = pd.DataFrame(data.failure) 
for col in range(len(dum_list)):
    dums = pd.get_dummies(data[dum_list[col]].astype('category'), prefix = dum_list[col], drop_first=True)
    merged_data_dums = merged_data_dums.join(dums)
merged_data_dums = merged_data_dums.drop('failure',axis=1)
x = merged_data_dums

#Add numerical features 
X = pd.DataFrame(data[['attribute1','attribute2','attribute4','attribute6']])
x = x.join(X)
```

```python
#Train Test Split w/ stratification
from sklearn.model_selection import train_test_split
y = pd.DataFrame(data.failure)
x_train,x_test, y_train, y_test = train_test_split(x,y,train_size = 0.6,random_state=0,stratify=y)

#Random Forest Classification
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf = clf.fit(x_train,y_train)
pred = clf.predict(x_test)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
conf = pd.DataFrame(confusion_matrix(y_test,pred),columns=['good','fail'])
conf 
```

```python
#Recall - What percentage of failures were caught?
from sklearn.metrics import recall_score
prec = recall_score(y_test,pred,labels=None,pos_label=1,average=None,sample_weight=None)
prec
```

```python
#Precision - What percentage of failure predictions were correct?
from sklearn.metrics import precision_score
rec = precision_score(y_test,pred,labels=None,pos_label=1,average=None,sample_weight=None)
rec
```

```python
#Feature Importance - Baseline Model
features = pd.DataFrame()
feat_comp = pd.DataFrame(columns=['Baseline','Downsample','SMOTE'])

features['Feature'] = x_train.columns
features['Importance'] = clf.feature_importances_
features.sort_values(by=['Importance'],ascending=True,inplace=True)
features.set_index('Feature',inplace=True)
features = features.tail(10)
features.plot(kind='barh',figsize=(20,20),fontsize=30)

feat_comp.Baseline = features.index
```

# Model w/ Downsampling

```python
#Isolate fail class data
data_fail = data[data.failure == '1']
#print(len(data_fail)) = 101

#Isolate good class data
data_good = data[data.failure == '0']
#print(len(data_good)) = 1067

#Select random sample of good class data
data_good = data_good.sample(frac = 0.1)

#Recombine data
Data = data_good.append(data_fail)
```

```python
#Dummy Categoricals
dum_list = ['attribute3','attribute5','attribute7','attribute9']
merged_data_dums = pd.DataFrame(Data.failure) 
for col in range(len(dum_list)):
    dums = pd.get_dummies(data[dum_list[col]].astype('category'), prefix = dum_list[col], drop_first=True)
    merged_data_dums = merged_data_dums.join(dums)
merged_data_dums = merged_data_dums.drop('failure',axis=1)
x = merged_data_dums

#Add numerical features 
X = pd.DataFrame(Data[['attribute1','attribute2','attribute4','attribute6']])
x = x.join(X)
```

```python
#Train Test Split w/ stratification
y = Data.failure
x_train,x_test, y_train, y_test = train_test_split(x,y,train_size = 0.6,random_state=0,stratify=y)

#Random Forest Classification
clf = RandomForestClassifier(n_estimators=100)
clf = clf.fit(x_train,y_train)
pred = clf.predict(x_test)
```

```python
#Confusion Matrix
conf = pd.DataFrame(confusion_matrix(y_test,pred),columns=['good','fail'])
conf 
```

```python
#Recall - What percentage of failures were caught?
prec = recall_score(y_test,pred,labels=None,pos_label=1,average=None,sample_weight=None)
prec
```

```python
#Precision - What percentage of failure predictions were correct?
rec = precision_score(y_test,pred,labels=None,pos_label=1,average=None,sample_weight=None)
rec
```

```python
#Feature Importance - Model w/ Downsampling
features = pd.DataFrame()
features['Feature'] = x_train.columns
features['Importance'] = clf.feature_importances_
features.sort_values(by=['Importance'],ascending=True,inplace=True)
features.set_index('Feature',inplace=True)
features = features.tail(10)
features.plot(kind='barh',figsize=(20,20),fontsize=30)

feat_comp.Downsample = features.index
```

# Model w/ SMOTE

```python
#Dummy Categoricals
dum_list = ['attribute3','attribute5','attribute7','attribute9']
merged_data_dums = pd.DataFrame(data.failure) 
for col in range(len(dum_list)):
    dums = pd.get_dummies(data[dum_list[col]].astype('category'), prefix = dum_list[col], drop_first=True)
    merged_data_dums = merged_data_dums.join(dums)
merged_data_dums = merged_data_dums.drop('failure',axis=1)
x = merged_data_dums

#Add numerical features 
X = pd.DataFrame(data[['attribute1','attribute2','attribute4','attribute6']])
x = x.join(X)

#Train Test Split w/ stratification
y = data.failure
x_train,x_test, y_train, y_test = train_test_split(x,y,train_size = 0.6,random_state=0,stratify=y)
```

```python
#Oversampling w/ SMOTE - training data only
    #conda install -c glemaitre imbalanced-learn

from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 0, ratio = 1.0)

#Oversample training data
x_train_res, y_train_res = sm.fit_sample(x_train, y_train)
```

```python
#Random Forest Classification
clf = RandomForestClassifier(n_estimators=100)
clf = clf.fit(x_train_res,y_train_res)
pred = clf.predict(x_test)
```

```python
#Confusion Matrix
conf = pd.DataFrame(confusion_matrix(y_test,pred),columns=['good','fail'])
conf 
```

```python
#Recall - What percentage of failures were caught?
prec = recall_score(y_test,pred,labels=None,pos_label=1,average=None,sample_weight=None)
prec
```

```python
#Precision - What percentage of failure predictions were correct?
rec = precision_score(y_test,pred,labels=None,pos_label=1,average=None,sample_weight=None)
rec
```

```python
#Feature Importance - Model w/ SMOTE
features = pd.DataFrame()
features['Feature'] = x_train.columns
features['Importance'] = clf.feature_importances_
features.sort_values(by=['Importance'],ascending=True,inplace=True)
features.set_index('Feature',inplace=True)
features = features.tail(10)
features.plot(kind='barh',figsize=(20,20),fontsize=30)

feat_comp.SMOTE = features.index
```
