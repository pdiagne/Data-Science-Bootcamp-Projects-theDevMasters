# Similar Products -  Co-occurrence Matrix

```python
#Setup data for co-occurrence matrix
orders = pd.DataFrame(data['Orders.id'])
products = pd.DataFrame(data['Products.id']).fillna(value=0).astype(int)
Data = products.join(orders)
Data.head()
```

```python
#https://stackoverflow.com/questions/31331310/pandas-combine-rows-that-share-an-association

import itertools as IT
edges = Data.groupby(['Orders.id'], group_keys=False).apply(
    lambda x: pd.DataFrame(list(IT.combinations(x['Products.id'], 2)), 
                           columns=['first', 'second'])).reset_index(drop=True)
edges.head()
```

```python
cooccurrence = pd.crosstab(index=[edges['first']], columns=[edges['second']])
items = Data['Products.id'].unique()
cooccurrence = cooccurrence.reindex(index=items, columns=items)
cooccurrence = cooccurrence.add(cooccurrence.T, fill_value=0)
cooccurrence = cooccurrence.fillna(0)
diagvals = cooccurrence.sum(axis=0)
arr = cooccurrence.values
i = np.arange(len(diagvals))
arr[i,i] = diagvals
cooccurrence = pd.DataFrame(arr, columns=cooccurrence.columns,index=cooccurrence.index)
cooccurrence.head(25)
```

```python
#Select similar products based on Co-occurrence Matrix
prod = 1430 #product number as integer (1430, 65, 100)
col = cooccurrence.loc[prod]
indices = np.nonzero(col)
similar = col.iloc[indices].sort_values(ascending=False)
similar
```
